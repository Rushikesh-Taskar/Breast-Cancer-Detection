```python
# drop all row having more than 2 null values

df.dropna(thresh=4, inplace=True)
df
```

```python
df['mean_area'].fillna(np.pi*df['mean_radius']*df['mean_radius'], inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df['mean_radius'].fillna(method='bfill', inplace=True)
df['mean_texture'].fillna(method='ffill', inplace=True)
df['mean_area'].fillna(method='bfill', inplace=True)
df['mean_smoothness'].fillna(method='ffill', inplace=True)
df
```

Decision Tree Model

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import matplotlib 
matplotlib.use('Agg')
import matplotlib.pyplot as plt
```

```python
xx = df.drop('diagnosis',axis=1)
yy = df['diagnosis'] # y test 

X_train, X_test, y_train, y_test = train_test_split(xx,yy,test_size=0.2)
dtree = DecisionTreeClassifier()
dtree = dtree.fit(X_train,y_train)

fig = plt.figure(figsize=(50,30))

tree.plot_tree(dtree, feature_names=xx.columns,filled=True,class_names=['0','1'])

fig.savefig("decistion_tree.png")
```

```python
y_pre=dtree.predict(X_test) #predict
y_pre
```

```python
cm_decision = np.array(confusion_matrix(y_test, y_pre, labels=[1,0]))
confusion = pd.DataFrame(cm_decision, index=['is_cancer', 'is_healthy'], columns=['predicted_cancer', 'predicted_healthy'])
confusion
```

```python
sns.heatmap(confusion, annot=True)
```

```python
print(classification_report(y_test, y_pre))
dt_Acc = accuracy_score(y_pre, y_test)
print('Decision Tree accuracy: {:.2f}%'.format(dt_Acc*100))
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(dtree, X=X_train, y=y_train, cv=10)
scores
```

```python
plt.plot(scores,marker='o')
```

```python
compare = pd.DataFrame({'Model': ['SVM', 'SVM (Scaled Data)', 'Decision Tree'], 
                        
    'Accuracy': [SVM_Acc*100, SVM_scaled_Acc*100, dt_Acc*100]})

compare.sort_values(by='Accuracy', ascending=False)
```

```python
colors = ['red','green','blue']
plt.figure(figsize=(5,3))
plt.title("barplot Represent Accuracy of different models")
plt.xlabel("Accuracy %")
plt.ylabel("Algorithms")
plt.bar(compare['Model'],compare['Accuracy'],color = colors)
plt.xticks(rotation=90)
plt.show()
```
